```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5']
students = [23,22,23,20,24]
ax.bar(langs,students)
ax.set_ylabel("Value of importance")
plt.autoscale()
plt.savefig("Bad_interpretation.png",bbox_inches='tight',dpi=1000)
plt.show()
```
