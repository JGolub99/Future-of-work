```python
import pandas as pd

df1 = pd.read_excel("unknown_jobs.xlsx")
df1.head()
```

```python
quant1 = 0.8 #Need a very tight bound for mathematics
quant2 = 0.5 #Can afford a less tight bound for the other skills
df2 = df1[(df1["Mathematics"] >= df1["Mathematics"].quantile(q=quant1))&(df1["Systems Analysis"] >= df1["Systems Analysis"].quantile(q=quant2))&(df1["Quality Control Analysis"] <= df1["Quality Control Analysis"].quantile(q=1-quant2))&(df1["Service Orientation"] <= df1["Service Orientation"].quantile(q=1-quant2))]
df2
```

```python
df2.to_csv("subsetjobs.csv",index=False)
```
