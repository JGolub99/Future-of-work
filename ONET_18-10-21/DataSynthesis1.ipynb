```python
''' In this doc we shall generate our own binary classifier '''
import numpy as np
import pandas as pd
```

```python
#Begin with ouput = 0
m = 4 #no. variables
n = 100  #no. observations

data_matrix0 = np.zeros(shape=(n,m))
data_matrix0[:,0] = np.random.uniform(0.0,0.5,n)
data_matrix0[:,1] = np.random.uniform(0.0,0.5,n) + np.random.normal(0.0,0.2,n)
data_matrix0[:,2] = np.random.uniform(0.0,0.5,n) + np.random.normal(0.0,0.5,n)
data_matrix0[:,3] = np.random.uniform(0.0,1.0,n)
data_matrix0 = np.c_[data_matrix0, np.zeros(n)]  
```

```python
data_matrix1 = np.zeros(shape=(n,m))
data_matrix1[:,0] = np.random.uniform(0.5,1.0,n)
data_matrix1[:,1] = np.random.uniform(0.5,1.0,n) + np.random.normal(0.0,0.2,n)
data_matrix1[:,2] = np.random.uniform(0.5,1.0,n) + np.random.normal(0.0,0.5,n)
data_matrix1[:,3] = np.random.uniform(0.0,1.0,n)
data_matrix1 = np.c_[data_matrix1, np.ones(n)]
```

```python
data_matrix = np.concatenate((data_matrix0, data_matrix1), axis=0)
```
